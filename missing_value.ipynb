```python
import pandas as pd
import missingno as msno
```

```python
titanic = pd.read_csv("data/titanic/train.csv")
print(titanic.shape)
titanic.head(3)
```

```python
msno.matrix(titanic, color = (0.4, 0.7, 1), labels=True)
# 左側縱坐標上的兩個數值為樣本數量的始末
# 共有891個樣本；12列中有10列含有缺失值
# 白線越多，缺失值越多
```

```python
msno.bar(titanic, color="#66B3FF") # , sort='descending'
# 左側縱坐標為數據非缺失值所佔的百分比
# 右側縱坐標為數據非缺失值的數量
```

```python
msno.heatmap(titanic)
# 缺失值的特徵之間的相關性，某特徵存在缺失值時另一特徵也會存在缺失值的相關程度。
# 沒有數字，代表特徵之間不存在缺失的相關性
# 數字為1，代表其中一個特徵缺失，另一個就一定會缺失
```
