```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
titanic = pd.read_csv("data/titanic/train.csv")
print(titanic.shape)
titanic.head(3)
```

### 類別變數對類別Target

```python
plt.figure(figsize = (3, 2))
sns.countplot(data = titanic, x = "Survived").set(title = "Survived")
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.countplot(data = titanic, x = 'Pclass', hue = "Survived")
plt.show()
```

```python
cat_col = ["Survived", "Pclass", "Sex", "Embarked"]
a, b = 2, 2
k = 0
fig, ax = plt.subplots(a, b, figsize = (10, 5))
for i in range(a):
    for j in range(b):
        sns.countplot(data = titanic, x = cat_col[k], hue = "Survived", ax = ax[i, j])
        ax[i, j].set(xlabel = None, ylabel = None, title = cat_col[k])
        k += 1
```

### 連續變數對類別Target

```python
plt.figure(figsize = (5, 3))
sns.boxplot(data = titanic, x = 'Age')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.boxplot(data = titanic, x = 'Survived', y = "Age")
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.boxplot(data = titanic, x = 'Pclass', y = "Age", hue = "Survived")
plt.show()
```

```python
num_col = ['Age', 'SibSp', 'Parch', 'Fare']
a, b = 2, 2
k = 0
fig, ax = plt.subplots(a, b, figsize = (10, 6))
for i in range(a):
    for j in range(b):
        sns.boxplot(x = "Pclass", y = num_col[k], hue = "Survived", data = titanic, width = 0.3, ax = ax[i, j])
        ax[i, j].set(xlabel = None, ylabel = None, title = num_col[k])
        k += 1
```

#### 連續變數自我分布

```python
num_col = ['Age', 'SibSp', 'Parch', 'Fare']
a, b = 2, 2
k = 0
fig, ax = plt.subplots(a, b, figsize = (10, 6))
for i in range(a):
    for j in range(b):
        sns.kdeplot(data = titanic, x = num_col[k], fill = True, ax = ax[i, j])
        ax[i, j].set(xlabel = None, ylabel = None, title = num_col[k])
        k += 1
```
